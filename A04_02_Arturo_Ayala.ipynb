```python
# Import necessary libraries for numerical operations and plotting
import numpy as np
import matplotlib.pyplot as plt

# --- Initial Data --- #
# Define initial data points for x and y
x_initial = np.array([10, 2, 3])
y_initial = np.array([2847, -4017, -1255])

# --- Search Interval Definition --- #
# Define the lower and upper bounds for the search interval
x_search_min = 2
x_search_max = 15

# --- Display Initial Data --- #
print("Initial Data Points:")
for i in range(len(x_initial)):
    print(f"x = {x_initial[i]:<4} | y = {y_initial[i]:>6}")

# --- Identify Current Best Point --- #
# Find the index of the minimum y-value
idx_best = np.argmin(y_initial)
x_current_best = x_initial[idx_best]
y_current_best = y_initial[idx_best]

print("\nCurrent Best Value Found:")
print(f"x_min = {x_current_best}")
print(f"y_min = {y_current_best}")

# --- Plot Initial Data --- #
plt.figure(figsize=(9, 6))
plt.scatter(x_initial, y_initial, s=100, color='blue', zorder=5, label='Initial Sample Points')
plt.title("Initial Sampled Points")
plt.xlabel("X-value")
plt.ylabel("Y-value")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()

# --- Generate Candidate Points within Interval --- #
# Create a range of candidate x-values within the defined interval
x_candidates_all = np.arange(x_search_min, x_search_max + 1, 1)

print("\nAll Candidate Points within Search Interval:")
print(x_candidates_all)

# --- Filter Out Already Sampled Points --- #
# Remove candidate points that have already been sampled
x_new_candidates = [val for val in x_candidates_all if val not in x_initial]
x_new_candidates = np.array(x_new_candidates)

print("\nCandidate Points Not Yet Sampled:")
print(x_new_candidates)

# --- Select Next Point Based on Proximity to Current Best --- #
# Calculate distances from new candidates to the current best x-value
distances_to_best = np.abs(x_new_candidates - x_current_best)
# Find the candidate closest to the current best x-value
idx_next_point = np.argmin(distances_to_best)
x_next_suggested = x_new_candidates[idx_next_point]

print("\nNext Suggested Point for Sampling:")
print(f"x_next = {x_next_suggested}")

# --- Plotting with Next Suggested Point --- #
plt.figure(figsize=(9, 6))
plt.scatter(x_initial, y_initial, s=100, color='blue', zorder=5, label='Initial Sample Points')
plt.axvline(x=x_next_suggested, color='red', linestyle='--', linewidth=2, label='Next X-value Suggested')
plt.scatter(x_current_best, y_current_best, s=150, color='green', marker='*', zorder=6, label='Current Best X-value')
plt.title("Selection of Next Point for Sampling (Closest to Current Best)")
plt.xlabel("X-value")
plt.ylabel("Y-value")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# --- Reproducibility Setup --- #
# Set a random seed for NumPy to ensure reproducible results
np.random.seed(42)

# --- Define the True (Objective) Function --- #
def f_objective(x):
    """The actual function we are trying to optimize (minimize)."""
    return np.sin(x) + 0.5 * x

# --- Define Search Interval and Grid --- #
# Set the bounds for the search space
x_interval_min = 2
x_interval_max = 15

# Define the number of points to generate for evaluation
num_grid_points = 10001

# Generate a finely spaced grid of x-values within the interval
x_grid = np.linspace(x_interval_min, x_interval_max, num_grid_points)

# --- Evaluate Function and Add Noise --- #
# Evaluate the true function over the grid
y_true = f_objective(x_grid)

# Add Gaussian noise to simulate real-world observations
# This creates 'noisy' observations of the true function
noise_std_dev = 0.5
epsilon_noise = np.random.normal(0, noise_std_dev, size=num_grid_points)
y_noisy_observations = y_true + epsilon_noise

# --- Find Minimum in Noisy Observations --- #
# Locate the index of the minimum y-value in the noisy data
idx_min_noisy = np.argmin(y_noisy_observations)
# Extract the x and y values corresponding to this minimum
x_best_found = x_grid[idx_min_noisy]
y_min_found = y_noisy_observations[idx_min_noisy]

print(f"Best x-value found from noisy data: {x_best_found:.4f}")
print(f"Minimum y-value found from noisy data: {y_min_found:.4f}")

# --- Visualization of Function and Minimum --- #
plt.figure(figsize=(10, 6))

# Plot the true underlying function
plt.plot(x_grid, y_true, label="True Objective Function", color='blue', linewidth=2)
# Scatter plot of noisy observations
plt.scatter(x_grid, y_noisy_observations, s=5, alpha=0.3, color='gray', label="Noisy Observations")
# Highlight the minimum point found in the noisy data
plt.scatter(x_best_found, y_min_found, color='red', marker='X', s=150, zorder=5, label="Global Minimum (Noisy Data)")

plt.title("Objective Function with Noise and Identified Minimum")
plt.xlabel("X-value")
plt.ylabel("Y-value")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Install the scikit-learn library if not already installed
# This is done quietly (-q) to prevent verbose output during installation
!pip -q install scikit-learn

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C

# --- Reproducibility Setup --- #
# Set a random state for reproducibility of results (e.g., kernel initialization)
random_state_seed = 42
np.random.seed(random_state_seed)

# --- Initial Observed Data --- #
# Define initial data points (X, y) that have been sampled
# X needs to be a 2D array for scikit-learn models
X_observed = np.array([10, 2, 3, 7, 12]).reshape(-1, 1)
y_observed = np.array([2847, -4017, -1255, 1773, 563])

print("Initial Observed Data Points:")
for i in range(len(X_observed)):
    print(f"x = {X_observed[i,0]:<4} | y = {y_observed[i]:>7}")

# --- Define Search Grid --- #
# Set the bounds for the x-axis search space
x_grid_min = 2
x_grid_max = 15
num_grid_points = 10001 # Number of points to evaluate the GPR over

# Create a finely spaced grid of x-values for prediction
X_test_grid = np.linspace(x_grid_min, x_grid_max, num_grid_points).reshape(-1, 1)

print("\nFirst 5 values of the prediction grid (X_test_grid):")
print(X_test_grid[:5].ravel())

print("\nLast 5 values of the prediction grid (X_test_grid):")
print(X_test_grid[-5:].ravel())

# --- Gaussian Process Regressor (GPR) Setup --- #
# Define the kernel for the Gaussian Process. Here, a ConstantKernel multiplied by an RBF kernel.
# C(1.0): A constant kernel, allowing the amplitude of the RBF kernel to be learned.
# RBF(length_scale=1.0): Radial Basis Function kernel, defining the smoothness.
kernel_gpr = C(1.0) * RBF(length_scale=1.0)

# Initialize the GPR model with the chosen kernel and random state
gpr_model = GaussianProcessRegressor(kernel=kernel_gpr, random_state=random_state_seed)

# --- Model Training --- #
# Fit the GPR model to the initial observed data (X_observed, y_observed)
gpr_model.fit(X_observed, y_observed)

# --- Prediction and Uncertainty Estimation --- #
# Predict mean (y_hat) and standard deviation (sigma) over the test grid
y_mean_predicted, y_std_dev = gpr_model.predict(X_test_grid, return_std=True)

# --- Calculate Confidence Bounds --- #
# Define upper and lower confidence bounds (e.g., 95% confidence interval using 2*sigma)
y_upper_bound = y_mean_predicted + 2 * y_std_dev
y_lower_bound = y_mean_predicted - 2 * y_std_dev

# --- Identify Estimated Minimum --- #
# Find the index of the minimum predicted mean value
idx_estimated_min = np.argmin(y_mean_predicted)
x_estimated_best = X_test_grid[idx_estimated_min, 0]
y_estimated_min = y_mean_predicted[idx_estimated_min]

print("\nEstimated X-value of the minimum:")
print(f"{x_estimated_best:.4f}")

print("\nEstimated Y-value of the minimum:")
print(f"{y_estimated_min:.4f}")

# --- Visualization of GPR Results --- #
plt.figure(figsize=(12, 7))

# Plot initial observed data points
plt.scatter(X_observed, y_observed, s=100, color='red', zorder=5, label="Initial Observed Data")
# Plot the predicted mean function
plt.plot(X_test_grid, y_mean_predicted, color='blue', label="GPR Mean Prediction")
# Fill the area between the confidence bounds to show uncertainty
plt.fill_between(X_test_grid.ravel(), y_lower_bound, y_upper_bound, color='blue', alpha=0.2, label="95% Confidence Interval")
# Highlight the estimated minimum point from the GPR mean prediction
plt.scatter(x_estimated_best, y_estimated_min, color='green', marker='*', s=200, zorder=6, label="Estimated Minimum Point")

plt.xlabel("X-value")
plt.ylabel("Y-value")
plt.title("Gaussian Process Regression: Mean Prediction and Uncertainty")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
# Install the scikit-learn library if not already installed
# This is done quietly (-q) to prevent verbose output during installation
!pip -q install scikit-learn

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF

# --- Reproducibility Setup --- #
# Set a random state for reproducibility of results
random_state_seed = 42
np.random.seed(random_state_seed)

# --- Initial Observed Data --- #
# Define initial data points (X, y) that have been sampled
# X needs to be a 2D array for scikit-learn models
X_sampled = np.array([10, 2, 3, 7, 12]).reshape(-1, 1)
y_sampled = np.array([2847, -4017, -1255, 1773, 563], dtype=float)

print("Initial Sampled Data:")
for i in range(len(X_sampled)):
    print(f"x = {X_sampled[i, 0]:<4} | y = {y_sampled[i]:>7.2f}")

# --- Define Search Interval and Grid --- #
# Set the bounds for the x-axis search space
x_search_min = 2
x_search_max = 15
num_grid_points = 10001 # Number of points to evaluate the GPR over

# Create a finely spaced grid of x-values for prediction
X_test_grid = np.linspace(x_search_min, x_search_max, num_grid_points).reshape(-1, 1)

# --- Gaussian Process Regressor (GPR) Setup --- #
# Define the kernel for the Gaussian Process (Radial Basis Function kernel)
kernel_gpr = RBF(length_scale=1.0)

# Initialize the GPR model with the chosen kernel and random state
gpr_model = GaussianProcessRegressor(kernel=kernel_gpr, random_state=random_state_seed)

# --- Model Training --- #
# Fit the GPR model to the initial observed data
gpr_model.fit(X_sampled, y_sampled)

# --- Prediction and Uncertainty Estimation --- #
# Predict mean (y_mean_predicted) and standard deviation (y_std_dev) over the test grid
y_mean_predicted, y_std_dev = gpr_model.predict(X_test_grid, return_std=True)

# --- Calculate Confidence Bounds (e.g., 2 standard deviations) --- #
y_upper_confidence = y_mean_predicted + 2 * y_std_dev
y_lower_confidence = y_mean_predicted - 2 * y_std_dev

# --- Acquisition Function: Upper Confidence Bound (UCB) for Exploration --- #
# In Bayesian Optimization for minimization, a common strategy for exploration
# is to sample where the upper confidence bound is maximized. This balances
# exploring uncertain regions (high std_dev) and potentially good regions (high mean).
# Here, we aim to find the maximum of y_upper_confidence to select the next point.
idx_next_sample = np.argmax(y_upper_confidence)
x_next_sample = X_test_grid[idx_next_sample, 0]
# The predicted mean value at this new sample point
y_predicted_at_next_sample = y_mean_predicted[idx_next_sample]

print("\nNext Suggested Point for Sampling (based on UCB acquisition):")
print(f"x_new = {x_next_sample:.4f}")
print(f"y_predicted_at_x_new = {y_predicted_at_next_sample:.2f}")

# --- Update Sampled Data --- #
# Add the newly suggested point to the historical sampled data
X_sampled_updated = np.vstack([X_sampled, [[x_next_sample]]])
y_sampled_updated = np.append(y_sampled, y_predicted_at_next_sample)

print("\nUpdated Sampled Data (including the new point):")
for i in range(len(X_sampled_updated)):
    print(f"x = {X_sampled_updated[i, 0]:<4} | y = {y_sampled_updated[i]:>7.2f}")

# --- Visualization of Bayesian Optimization Step --- #
plt.figure(figsize=(12, 7))

# Plot the initial observed data points
plt.scatter(X_sampled, y_sampled, s=100, color='red', zorder=5, label="Initial Sampled Points")
# Plot the GPR mean prediction
plt.plot(X_test_grid, y_mean_predicted, color='blue', label="GPR Mean Prediction (\u03BC)")
# Plot the upper confidence bound (acquisition function for exploration)
plt.plot(X_test_grid, y_upper_confidence, color='green', linestyle='--', label="Upper Confidence Bound (\u03BC + 2\u03C3)")
# Plot the lower confidence bound
plt.plot(X_test_grid, y_lower_confidence, color='purple', linestyle='--', label="Lower Confidence Bound (\u03BC - 2\u03C3)")
# Highlight the newly suggested point for sampling
plt.scatter(x_next_sample, y_predicted_at_next_sample, s=200, color='orange', marker='X', zorder=6, label="Next Point to Sample (UCB Max)")

plt.xlabel("X-value")
plt.ylabel("Y-value")
plt.title("Bayesian Optimization: GPR with Upper Confidence Bound Acquisition")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
# Install the scikit-learn library if not already installed
# This is done quietly (-q) to prevent verbose output during installation
!pip -q install scikit-learn

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF

# --- Reproducibility Setup --- #
# Set a random state for reproducibility of results
random_state_seed = 42
np.random.seed(random_state_seed)

# --- Initial Observed Data --- #
# Define initial data points (X, y) that have been sampled
# X needs to be a 2D array for scikit-learn models
X_initial_samples = np.array([10, 2, 3, 7, 12]).reshape(-1, 1)
y_initial_samples = np.array([2847, -4017, -1255, 1773, 563], dtype=float)

# --- Define Search Grid --- #
# Create a finely spaced grid of x-values for prediction
X_test_grid = np.linspace(2, 15, 10001).reshape(-1, 1)

# --- Gaussian Process Regressor (GPR) Setup --- #
# Define the kernel for the Gaussian Process (Radial Basis Function kernel)
kernel_gpr = RBF(length_scale=1.0)

# Initialize the GPR model with the chosen kernel and random state
gpr_model = GaussianProcessRegressor(kernel=kernel_gpr, random_state=random_state_seed)

# --- Model Training --- #
# Fit the GPR model to the initial observed data
gpr_model.fit(X_initial_samples, y_initial_samples)

# --- Prediction and Uncertainty Estimation --- #
# Predict mean (y_mean_predicted) and standard deviation (y_std_dev) over the test grid
y_mean_predicted, y_std_dev = gpr_model.predict(X_test_grid, return_std=True)

# --- Calculate Confidence Bounds (e.g., 1 standard deviation) --- #
y_upper_bound = y_mean_predicted + y_std_dev
y_lower_bound = y_mean_predicted - y_std_dev

# --- Acquisition Function: Upper Confidence Bound (UCB) for Exploration --- #
# Identify the point where the upper confidence bound is maximized
# This point represents a balance between exploration (high uncertainty) and exploitation (good mean)
idx_max_ucb = np.argmax(y_upper_bound)
x_new_suggested = X_test_grid[idx_max_ucb, 0]
# The predicted mean value at this new suggested point
y_predicted_at_new_x = y_mean_predicted[idx_max_ucb]

# --- Display Results --- #
print("Initial Sampled Data Points:")
for i in range(len(X_initial_samples)):
    print(f"x = {X_initial_samples[i, 0]:<4} | y = {y_initial_samples[i]:>7.2f}")

print("\nNext Suggested Point for Sampling:")
print(f"x_new = {x_new_suggested:.4f}")
print(f"y_predicted_at_x_new = {y_predicted_at_new_x:.2f}")

# --- Update Data with New Suggested Point --- #
# This simulates adding the new point to the dataset after it has been sampled
X_updated_samples = np.vstack([X_initial_samples, [[x_new_suggested]]])
y_updated_samples = np.append(y_initial_samples, y_predicted_at_new_x)

print("\nUpdated Sampled Data (including the new suggested point):")
for i in range(len(X_updated_samples)):
    print(f"x = {X_updated_samples[i, 0]:<4} | y = {y_updated_samples[i]:>7.2f}")

# --- Visualization of GPR and Next Sample Point --- #
plt.figure(figsize=(12, 7))

# Plot the initial sampled data points (excluding the newly added one for clarity)
plt.scatter(X_initial_samples[:, 0], y_initial_samples, s=100, color='red', zorder=5, label="Initial Sampled Points")
# Plot the predicted mean function from the GPR model
plt.plot(X_test_grid[:, 0], y_mean_predicted, color='blue', label="GPR Mean Prediction (\u03BC)")
# Plot the upper confidence bound (acquisition function)
plt.plot(X_test_grid[:, 0], y_upper_bound, color='green', linestyle='--', label="Upper Confidence Bound (\u03BC + \u03C3)")
# Plot the lower confidence bound
plt.plot(X_test_grid[:, 0], y_lower_bound, color='purple', linestyle='--', label="Lower Confidence Bound (\u03BC - \u03C3)")
# Highlight the newly suggested point for sampling
plt.scatter(x_new_suggested, y_predicted_at_new_x, s=200, color='orange', marker='X', zorder=6, label="Next Point to Sample (UCB Max)")

plt.xlabel("X-value")
plt.ylabel("Y-value")
plt.title("Gaussian Process Regression: Mean Prediction and 1-Sigma Bounds with Next Sample Selection")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
