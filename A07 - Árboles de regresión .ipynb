```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.tree import DecisionTreeRegressor
```

```python
df= pd.read_csv('Advertising.csv')
```

```python
X = df[['TV', 'radio', 'newspaper']]
y= df['sales']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

lr = LinearRegression()
lr.fit(X_train, y_train)

r2_lr_train = r2_score(y_train, lr.predict(X_train))
r2_lr_test = r2_score(y_test, lr.predict(X_test))
```

```python
dt = DecisionTreeRegressor(random_state=42)
dt.fit(X_train, y_train)

r2_dt_train = r2_score(y_train, dt.predict(X_train))
r2_dt_test = r2_score(y_test, dt.predict(X_test))
```

```python
print("--- Resultados del R^2 ---")
print(f"Regresión Lineal (train): {r2_lr_train:}")
print(f"Árbol de Decisión (train): {r2_dt_train:}")
print(f"Regresión Lineal (test): {r2_lr_test:}")
print(f"Árbol de Decisión (test): {r2_dt_test:}")
```
