En esta actividad vamos a explorar los datos y luego entrenaremos una Red Neuronal (Perceptrón Multicapa). Usaremos Optimización Bayesiana para encontrar la mejor combinación de neuronas, función de activación y el valor de alpha, justo como se vio en clase.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier

# Cargar el dataset
df = pd.read_csv("Heart Prediction Quantum Dataset.csv")

# ==========================================
# Paso 1: EDA
# ==========================================
print("--- Paso 1: Análisis Exploratorio (EDA) ---")
print(df.head())
print(df.describe())
print("-" * 50)

# Separar variables predictoras y variable objetivo
X = df.drop(columns=["HeartDisease"])
y = df["HeartDisease"]

# ==========================================
# Paso 3: Splitting data (30% test_size)
# (Se hace antes del preprocesamiento por buenas prácticas)
# ==========================================
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# ==========================================
# Paso 2: Preprocessing
# Usaremos StandardScaler
# ==========================================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ==========================================
# Paso 4: Modeling (MLPClassifier)
# ==========================================
# Usamos los parámetros
model = MLPClassifier(hidden_layer_sizes=(5, 10), random_state=42, max_iter=1000)
model.fit(X_train_scaled, y_train)

# ==========================================
# Paso 5: Scoring (accuracy)
# ==========================================
accuracy_train = model.score(X_train_scaled, y_train)
accuracy_test = model.score(X_test_scaled, y_test)

print(f"Accuracy en entrenamiento: {accuracy_train:.4f}")
print(f"Accuracy en prueba (test): {accuracy_test:.4f}")
```
