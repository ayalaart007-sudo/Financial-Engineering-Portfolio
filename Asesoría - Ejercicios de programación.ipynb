# Ejercicio 1: La serie de 7 juegos

```python
import numpy as np

N = 1_000_000

# Simulamos 6 juegos. Probabilidad de ganar = 0.5
# Esto nos dará el número de juegos que ganó el equipo A en esos 6 juegos.
juegos_ganados = np.random.binomial(n=6, p=0.5, size=N)

# Si ganó exactamente 3, la serie se va al séptimo juego
se_va_a_7 = np.mean(juegos_ganados == 3)

print(f"1) Probabilidad de irse a 7 juegos: {se_va_a_7:.4f} (Teórico: {5/16:.4f})")
```

# Ejercicio 2: La prueba de la enfermedad (Bayes)

```python
import numpy as np

N = 10_000_000 # Usamos 10 millones porque el profe dijo que es para mayor precisión porque la enfermedad es rara (1 en 1000)

# 1. Simulamos quién tiene la enfermedad (1 de cada 1000)
enfermos = np.random.binomial(n=1, p=1/1000, size=N)

# 2. Simulamos el resultado de la prueba
# Si está enfermo (p=0.98 de dar positivo)
prueba_si_enfermo = np.random.binomial(n=1, p=0.98, size=N)
# Si no está enfermo (p=0.01 de dar positivo / falso positivo)
prueba_si_sano = np.random.binomial(n=1, p=0.01, size=N)

# Asignamos el resultado de la prueba dependiendo de si está enfermo o sano
prueba_positiva = np.where(enfermos == 1, prueba_si_enfermo, prueba_si_sano)

# 3. Filtramos SOLO los casos donde la prueba dio positiva
casos_positivos = (prueba_positiva == 1)

# Calculamos la probabilidad: De los que dieron positivo, ¿cuántos realmente tienen la enfermedad?
prob_realmente_enfermo = np.sum(enfermos[casos_positivos] == 1) / np.sum(casos_positivos)

print(f"2) Probabilidad de tener la enfermedad dado un positivo: {prob_realmente_enfermo:.4f} (Teórico: {98/1097:.4f})")
```

# Ejercicio 3: Los tres amigos y la lluvia

```python
import numpy as np

N = 1_000_000

# 1. Simulamos el clima real (25% de que llueva)
lluvia = np.random.binomial(n=1, p=0.25, size=N)

# 2. Simulamos lo que dicen los amigos.
# Si llueve, dirán que sí si dicen la verdad (prob 2/3). Necesitamos que los 3 digan que sí.
dicen_si_lluvia = np.random.binomial(n=3, p=2/3, size=N)

# Si NO llueve, dirán que sí solo si mienten (prob 1/3).
dicen_si_nolluvia = np.random.binomial(n=3, p=1/3, size=N)

# Resultado de lo que dijeron los 3 amigos
amigos_dicen_si = np.where(lluvia == 1, dicen_si_lluvia, dicen_si_nolluvia)

# 3. Nos quedamos solo con los universos donde los 3 amigos dijeron que SÍ llovía (valor = 3)
casos_dicen_si = (amigos_dicen_si == 3)

# Calculamos: De las veces que los 3 dijeron que sí, ¿cuántas veces sí llovía?
prob_lluvia_real = np.sum(lluvia[casos_dicen_si] == 1) / np.sum(casos_dicen_si)

print(f"3) Probabilidad de que sí esté lloviendo: {prob_lluvia_real:.4f} (Teórico: {8/11:.4f})")
```
